```python
file_name = 'cleaned_data'
data = dm.read_data(file_name, header = 0)
data.head()
```

```python
df = data.copy()
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.shape
```

```python
df.drop(['Year of project', 'Estimated effort', 'Estimated duration']
        , axis = 1, inplace = True)
```

```python
#These are the missing values rates for each column in our dataset
nan_rates_cols = dv.nan_columns_rates(data)
nan_rates_cols[nan_rates_cols != 0]
```

```python
imputed_data = dp.NearestNeighborsImputation(df)
```

## Final Missing Values HeatMap

```python
plt.figure(figsize = (20, 20))
dv.missing_values_map(imputed_data)
plt.title('Missing values heatmap')
```

# Features Anomaly treatment

```python
imputed_data['Size of organization'] = imputed_data['Size of organization'].replace({17: 16})
imputed_data['Customer organization type'] = imputed_data['Customer organization type'].replace({17: 16})
imputed_data['Required reusability'] = imputed_data['Required reusability'].replace({5: 4})
```

# Treating Target Outliers 

```python
plt.figure()
sns.boxplot(x = imputed_data['Actual effort'])
```

```python
from scipy.stats import zscore
```

```python
imputed_data['Z_Score'] = np.abs(zscore(imputed_data['Actual effort']))
imputed_data['Z_Score'].sort_values(ascending = False)
```

```python
non_outliers_data = imputed_data[imputed_data['Z_Score'] < 6]
non_outliers_data.drop(['Z_Score'], axis = 1, inplace = True)
non_outliers_data.shape
```

```python
dm.save_data(non_outliers_data, 'famd_data')
```

# Dropping low variance Features

```python
variance = np.round(non_outliers_data.var().sort_values(ascending = False), 2)
variance[20:]
```

```python
non_outliers_data.drop(list(variance[variance < 0.3].index), axis = 1, inplace = True)
```

# Drop highly correlated features

```python
cat_cols = [col for col in dd.categorical_cols_names() if col in non_outliers_data.columns]
num_cols = [col for col in non_outliers_data.columns if col not in cat_cols]
```

```python
dv.corr_matrix(non_outliers_data[num_cols])
```

```python
non_outliers_data.drop(['Dedicated team members', 'Other sizing method', 'Analysts capability'], axis = 1, inplace = True)
```

```python
cat_cols = [col for col in dd.categorical_cols_names() if col in non_outliers_data.columns]
```

```python
categorical_features_association = dv.chi2_test_categorical_features(cat_cols, imputed_data)
```

```python
categorical_features_association = categorical_features_association[categorical_features_association < 0.05]
```

```python
import itertools
comb = []
for subset in itertools.combinations(categorical_features_association.columns, 2):
   comb.append(subset)
```

```python
len(non_outliers_data.columns)
```

```python
for com in comb:
    if com[0] in non_outliers_data.columns and com[1] in non_outliers_data.columns:
        if categorical_features_association[com[0]][com[1]] < 0.05:
            if dv.correlation_actualEffort_categoricalData([com[0]], non_outliers_data).values > dv.correlation_actualEffort_categoricalData([com[1]], non_outliers_data).values:
                non_outliers_data.drop(com[1], axis = 1, inplace = True)
            else:
                non_outliers_data.drop(com[0], axis = 1, inplace = True)
```

```python
non_outliers_data.columns.sort_values()
```

```python
categorical_columns = [col for col in dd.categorical_cols_names() if col in non_outliers_data.columns]
numerical_columns = [col for col in non_outliers_data.columns if col not in categorical_columns]
```

# Multicolinearity test

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = non_outliers_data[numerical_columns].drop('Actual effort', axis = 1)
vif_results = pd.Series(index = vif_data.columns)
for i in range(vif_data.shape[1]):
    vif_results[vif_data.columns[i]] = variance_inflation_factor(vif_data.values, i)
vif_results.sort_values(ascending = False)
```

```python
non_outliers_data.drop(vif_results[vif_results > 10].index, axis = 1, inplace = True)
non_outliers_data.columns.sort_values()
```

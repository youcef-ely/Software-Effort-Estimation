## Deleting spaces of column name

```python
data.columns
```

```python
columns = [column[:-1] if column[-1:] == ' ' else column for column in data.columns]
columns
```

```python
final_columns = []
for col in columns:
    col = col.split(' ')
    if '' in col:
        col.remove('')
    col = ' '.join(col)
    final_columns.append(col)
```

```python
final_columns
```

```python
data.columns = final_columns
```
